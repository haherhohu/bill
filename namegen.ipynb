korean name generator


```python
import random
import pickle

# number of names to generate
n_names = 100
# name length for each name
namelength = (3, 10)
```

```python
result = "\n".join([''.join([f.decode(encoding='utf-16', errors=u'ignore') 
                             for f in [e.to_bytes(3,'little') 
                                       for e in random.sample(range(44032,55203), 
                                                              random.randint(*namelength)) ]]) 
                                                              for _ in range(n_names)])

print(result)
```

```python
# 완성형 준비작업
ksx1001 = zip(range(0xB0A1, 0xC8A1, 0x0100), range(0xB0FE, 0xC8FE, 0x0100))
# 완성형 한글 타겟 설정
targets = [e for r in [range(s,e) for s, e in ksx1001] for e in r]

# 한줄로
result = "\n".join([''.join([e.decode(encoding='euc-kr', errors=u'ignore') 
                             for e in [e.to_bytes(2,'big') 
                                       for e in random.sample(targets, 
                                                              random.randint(*namelength))]]) 
                                                              for _ in range(n_names)])
print(result)
```

```python
# 전자가족관계시스템 등록 남녀 8014개 이름 (중복 제거)
first = list(pickle.load(open("first_names.pkl", "rb")))
# 한국의 주요 성씨 207개 (중복 제거)
last = list(pickle.load(open("last_names.pkl", "rb")))
# 하나의 스트링으로 합치고 중복 문자를 모두 제거
targets = set(" ".join("".join(first + last)).split(" "))
# remove '' element
targets.discard('')

# 굳이 랜덤으로 뽑아봄
result = "\n".join([''.join(random.sample(list(targets), 
                                          random.randint(*namelength))) 
                                          for _ in range(n_names)])

print(result)
```

```python
# 전자가족관계시스템 등록 남녀 8014개 이름 (중복 제거)
first = list(pickle.load(open("first_names.pkl", "rb")))
# 한국의 주요 성씨 207개 (중복 제거)
last = list(pickle.load(open("last_names.pkl", "rb")))
# 이건 역시 상식적
result = "\n".join(["".join(random.sample(last, 1) + random.sample(first, 1)) 
                    for _ in range(n_names)])

print(result)
```
